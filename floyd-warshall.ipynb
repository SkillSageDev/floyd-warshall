```python
import numpy as np
import random
import time
```

```python
def graphToMatrix(graph):
    keyIndex = {
        key: i for i, key in enumerate(graph.keys())
    }  # this is a variable that makes a dictionary by making a loop; to say that ex: A -> 0, B -> 1, C -> 2 & so on
    size = len(graph)  # size of single row, or single column
    a0 = np.full((size, size), 1000)
    for edge in graph:
        b = graph[edge]
        for directedEdge in b:
            i = keyIndex[edge]
            j = keyIndex[directedEdge]
            a0[i][j] = b[directedEdge]
        np.fill_diagonal(a0, 0)
    return a0
```

 Step-by-Step Explanation

1. Convert the graph into adjecency matirx
2. Initialize the distance from each node to it self as 0 
3. For every vertex k, treat it as intermediate node 
4. For every parir of vetices (i,j), check :
    distance [i][j] = min(distance[i][j], distance[i][k] + distance[k][j])
5. Repeat until all vertices have been used as  intermediate nodes
6. the final matrix contains the shortest distance between all pairs
 

```python
def floydWarshall(graph):
    a = graphToMatrix(graph)
    size = a.shape[0]  # size of single row or single column
    for k in range(0, size):
        for i in range(0, size):
            for j in range(0, size):
                if i == j or i == k or j == k:
                    continue
                else:
                    a[i, j] = min(a[i, j], a[i, k] + a[k, j])
        # print(a) shows the steps 
    return(a)   
        
    
```

```python
graph = {
    "A": {"B": 2, "D": 3},
    "B": {"A": 3, "C": 2},
    "C": {"D": 4},
    "D": {"A": -2, "B": 6},
}

floydWarshall(graph)
```

Time Complexity Analysis

Floyd-Warshell algorithm uses three nested loops:

- The outer loop runs n times
- The middle loop runs n times
- The inner loop runs n times

Time complexity :
n x n x n = O(n^3)





```python
#Dataset Generation
def generate_graph(n, max_weight=10):
    graph = {}
    for i in range(n):
        graph[str(i)] = {}
        for j in range(n):
            if i != j and random.random() < 0.3:
                graph[str(i)][str(j)] = random.randint(1, max_weight)
    return graph
```

```python
#Executuion Time Measurment
def measure_time(n):
    graph = generate_graph(n)
    start = time.time()
    floydWarshall(graph)
    end = time.time()
    elapsed = end - start
    print(f"Dataset size n={n} took {elapsed:.6f} seconds")

```

Note:

Since FloydWarshell Algorithm time complexity is O(n^3)

Large inputs such as n = 1000 will not run on standard hardware

So we will test using small inputs.

```python
#Time Tests
inputs = [10,20,50,100,200,300]

for n in inputs:
    measure_time(n)
```
